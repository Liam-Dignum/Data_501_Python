```python
filtered_by_year = gamesDF[(gamesDF['Year'] >= 2000) & (gamesDF['Year'] <= 2015)]
sales_by_year = filtered_by_year.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum().reset_index()
sales_by_year
melted_sales = pd.melt(sales_by_year, id_vars=['Year'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'],
                       var_name='Region', value_name='Sales')
melted_sales
```

```python
sns.lineplot(x='Year', y='Sales', hue='Region', data=melted_sales, marker='o')
```

```python
filtered_by_year = gamesDF[(gamesDF['Year'] >= 1980) & (gamesDF['Year'] <= 2015)]
sales_by_year = filtered_by_year.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum().reset_index()
sales_by_year
```

```python
plt.stackplot(sales_by_year.Year,sales_by_year.NA_Sales,sales_by_year.EU_Sales,sales_by_year.JP_Sales,sales_by_year.Other_Sales,labels=['NA', 'EU', 'JP', 'Other'], colors=['blue', 'green', 'red', 'purple'])
plt.xlabel('Year', fontsize=14)
plt.ylabel('Share of Global Sales', fontsize=14)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
sales_by_year['TotalSales'] = sales_by_year[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum(axis=1)
sales_by_year['NA_Share'] = sales_by_year['NA_Sales'] / sales_by_year['TotalSales']
sales_by_year['EU_Share'] = sales_by_year['EU_Sales'] / sales_by_year['TotalSales']
sales_by_year['JP_Share'] = sales_by_year['JP_Sales'] / sales_by_year['TotalSales']
sales_by_year['Other_Share'] = sales_by_year['Other_Sales'] / sales_by_year['TotalSales']
plt.figure(figsize=(12, 8))
plt.stackplot(sales_by_year['Year'], 
              sales_by_year['NA_Share'], 
              sales_by_year['EU_Share'], 
              sales_by_year['JP_Share'], 
              sales_by_year['Other_Share'],
              labels=['NA', 'EU', 'JP', 'Other'],
              colors=['blue', 'green', 'red', 'purple'])

plt.title('Share of Global Sales by Region Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Share of Global Sales', fontsize=14)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
